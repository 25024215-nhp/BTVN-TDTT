
### W3A1
Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ 
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử 
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý

```python
n = int(input())

if 100 <= n <= 999:
    a = n // 100
    b = (n // 10) % 10
    c = n % 10
    
    print(c * 100 + b * 10 + a)
else:
    print("Không xử lý")
```

### WBai them
Cho 1 số nguyên n, kiểm tra trong tháng n có bao nhiêu ngày. Biết năm là năm nhuận.

```python
n=int(input("Nhap thang: "))
match n:
    case 1|3|5|7|8|10|12:
       s="Thang nay co 31 ngay"
    case 4|6|9|11:
       s="Thang nay co 30 ngay"
    case 2:
       s="Thang nay co 29 ngay"
    case _:
       s="Thang nay khong co trong nam"
print(s)
```
